```python
import pandas as pd

df1=pd.DataFrame({
    'cabins':['3rd_left','3rd_right','4th_left','4th_right','5th_left','5th_right'],
    'strength':[80,120,80,120,80,120]
})

df2=pd.DataFrame({
    'cabins':['3rd_left','3rd_right','4th_right','5th_left','5th_right'],
    'temperature':[24,23,22,28,29],
    'strength':[120,80,120,80,120]
})
```

```python
df3=pd.merge(df1,df2,on='cabins',how='outer',indicator=True,suffixes=('_left','_right'))
df3
```
